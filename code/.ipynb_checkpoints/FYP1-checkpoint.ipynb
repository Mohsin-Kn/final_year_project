# **6. Split the dataset into training and testing subsets.**

```python
X = df[numerical_columns]  # Features
y = df['Class_Name']       # Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Get the total number of rows
total_rows = len(df)

# Calculate 25% of the total rows
training_data_rows = int(0.80 * total_rows)

# Print the result
print(f"Training data: 80% of {total_rows} = {training_data_rows}")


# Calculate 25% of the total rows
testing_data_rows = int(0.20 * total_rows)

# Print the result
print(f"Testing data: 20% of {total_rows} = {testing_data_rows}")
```

# **7. SVC**

```python
# Training an SVM with class weights
svc_model = SVC(kernel='linear', class_weight='balanced', random_state=42)
svc_model.fit(X_train, y_train)


```

## **7.1 Printing  Training and Testing Accuries of SVC**

```python
#Training accuracy
y_train_pred_svc = svc_model.predict(X_train)
train_accuracy_svc = accuracy_score(y_train, y_train_pred_svc)


# Testing accuracy
y_test_pred_svc = svc_model.predict(X_test)
test_accuracy_svc = accuracy_score(y_test, y_test_pred_svc)

# Print the accuracies
print(f"Training Accuracy: {train_accuracy_svc:.4f}")
print(f"Testing Accuracy: {test_accuracy_svc:.4f}")
```

## **7.2 Printing Classificaton Report**

```python
print(classification_report(y_test, y_test_pred_svc))
```

## **7.3 Ploting Training and Testing Accuries of SVC**

```python
import matplotlib.pyplot as plt

# Function to plot training and testing accuracy
def plot_accuracy(train_accuracy, test_accuracy, model_name):
    # Categories for the bar plot
    categories = ['Training Accuracy', 'Testing Accuracy']
    accuracies = [train_accuracy, test_accuracy]

    # Create the plot
    plt.figure(figsize=(8, 5))
    plt.bar(categories, accuracies, color=['blue', 'orange'], width=0.4)
    plt.ylim(0, 1)
    plt.title(f'{model_name} - Training vs Testing Accuracy', fontsize=14)
    plt.ylabel('Accuracy', fontsize=12)

    # Add values on top of the bars
    for i, acc in enumerate(accuracies):
        plt.text(i, acc + 0.02, f"{acc:.2f}", ha='center', fontsize=12)

    # Add grid lines
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python


train_accuracy_svm = train_accuracy_svc
test_accuracy_svm = test_accuracy_svc
plot_accuracy(train_accuracy_svc, test_accuracy_svc, 'SVC')
```

## **7.4 Confusion matrix for SVC**

```python
# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, model_name, classes):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

# Retrieve the class names from the label encoder
classes = label_encoder.classes_  # Ensure label_encoder is defined earlier
```

```python
# SVM model confusion matrix
plot_confusion_matrix(y_test, y_test_pred_svc, 'SVC', classes)
```

# **7.5 Displaying Errors of SVC**

```python
# Function to calculate and display errors
def calculate_errors(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f"\n{model_name} Errors:")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# SVM Errors
calculate_errors(y_test, y_test_pred_svc, "SVC")
```

# **9. LSTM**
128: Number of LSTM units (hidden neurons).
input_shape: (1, 9) (time_steps=1, features=9).<br>
activation='tanh': Activation function for LSTM cells.<br>
return_sequences=False: Output only the final time step.<br>
Dense(4): Final output layer has 4 neurons (equal to the number of classes).<br>
activation='softmax': Converts outputs into probabilities summing to 1.<br>
learning_rate=0.0005: Controls step size during weight updates.<br>
sparse_categorical_crossentropy: Used for multi-class classification when target labels are integers<br>
epochs=20: Train for 20 full passes over the dataset.<br>
batch_size=32: Process 32 samples at a time.<Br>
validation_data: (X_test_lstm, y_test) for validation.<br>


```python
# Step 1: Reshape the data for LSTM (samples, time_steps, features)
#(samples, time_steps, features)
X_train_lstm = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))  # Reshaping for LSTM
X_test_lstm = X_test.values.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Step 2: Build the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(128, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]), activation='tanh', return_sequences=False))
model_lstm.add(Dropout(0.2))
model_lstm.add(Dense(4, activation='softmax'))  # softmax activation --> Converts raw scores into probabilities

# Step 3: Compile the model
model_lstm.compile(optimizer=Adam(learning_rate=0.0005), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Step 4: Train the LSTM model
history = model_lstm.fit(X_train_lstm, y_train, epochs=20, batch_size=32, validation_data=(X_test_lstm, y_test))
```

## **9.1 Printing  Training and Testing Accuries of LSTM**

```python
# Step 5: Evaluate the model on training and testing data
train_accuracy_lstm = history.history['accuracy'][-1]
test_loss_lstm, test_accuracy_lstm = model_lstm.evaluate(X_test_lstm, y_test, verbose=0)

print(f"Training Accuracy: {train_accuracy_lstm:.4f}")
print(f"Testing Accuracy: {test_accuracy_lstm:.4f}")
```

## **9.2 Printing Classificaton Report of LSTM**

```python
# Step 6: Predict the classes for the test set
y_pred_lstm_probs = model_lstm.predict(X_test_lstm)  # Get predicted probabilities / Matrix of probabilities for each class
y_pred_lstm = np.argmax(y_pred_lstm_probs, axis=1)  #For each sample, selects the class with the highest probability.

# Step 7: Print the classification report
print("\nClassification Report for LSTM:")
print(classification_report(y_test, y_pred_lstm))
```

## **9.3 Ploting Training and Testing Accuries of LSTM**

```python
#Training and Testing Accuracy for LSTM
plot_accuracy(train_accuracy_lstm, test_accuracy_lstm, 'LSTM')
```
